```python
#genByCategory = generatorsFinal.groupby(['CLOSEST_RECYCLING_FACILITY'])[['CLOSEST_RECYCLING_FACILITY']].count()
genByCategoryPercentages = (generatorsFinal['GENERATOR_CATEGORY'].value_counts(normalize=True) * 100).to_frame()
genByCategoryPercentages
```

```python
genByCategoryPercentages = genByCategoryPercentages.reset_index()
genByCategoryPercentages = genByCategoryPercentages.rename(columns={'index':'GeneratorCategory','GENERATOR_CATEGORY':'PercentageIndustry'})
genByCategoryPercentages
```

```python
# palette
pastel_palette = sns.color_palette("pastel")

# sort descending order
genByCategoryPercentages = genByCategoryPercentages.sort_values(by='PercentageIndustry', ascending=False)

# plotting 
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(genByCategoryPercentages['PercentageIndustry'], autopct='%1.1f%%', 
                                   startangle=140, colors=pastel_palette, labels=None)

# Adding legend with 'GeneratorCategory' and 'PercentageShare' in descending order
legend_labels = [f"{cat}: {share:.1f}%" for cat, share in zip(genByCategoryPercentages['GeneratorCategory'], genByCategoryPercentages['PercentageIndustry'])]
plt.legend(wedges, legend_labels, title='Generator Category & Industry Dominance', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1), fontsize='large')

# Adding title
plt.title('Food Waste Generators by Industry Dominance')

# Display the plot
plt.show()
```

```python
genByCategoryPercentages = generatorsFinal['GENERATOR_CATEGORY'].count()
genByCategoryPercentages
```

```python
genByCategoryFoodWaste = generatorsFinal.groupby('GENERATOR_CATEGORY').agg({'EXCESSFOOD_TONYEAR_AVERAGE':'sum'}).sort_values(by = 'EXCESSFOOD_TONYEAR_AVERAGE', ascending = False)

genByCategoryFoodWaste = genByCategoryFoodWaste.reset_index()
genByCategoryFoodWaste = genByCategoryFoodWaste.rename(columns={'index':'GeneratorCategory'})
genByCategoryFoodWaste['PercentageShare'] = (genByCategoryFoodWaste['EXCESSFOOD_TONYEAR_AVERAGE'] / genByCategoryFoodWaste['EXCESSFOOD_TONYEAR_AVERAGE'].sum()) * 100
genByCategoryFoodWaste
```

```python
# Plotting a pie chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(genByCategoryFoodWaste['EXCESSFOOD_TONYEAR_AVERAGE'], labels=None, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))

# Adding legend with 'GENERATOR_CATEGORY' and 'Percentage'
legend_labels = [f"{cat}: {percentage:.1f}%" for cat, percentage in zip(genByCategoryFoodWaste['GENERATOR_CATEGORY'], genByCategoryFoodWaste['PercentageShare'])]
plt.legend(wedges, legend_labels, title='Generator Category & Percentage', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

# Adding title
plt.title('Excess Food Waste by Generator Category')

# Display the plot
plt.show()
```
